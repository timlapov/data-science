#### Обработка пропусков

Площадь квартиры – это основной параметр, который влияет на стоимость квадратного метра. Если мы заполним пропуски средним/медианным значением, то это може сильно повлиять на стоимость квадратного метра в небольших населенных пунктах. Поэтому мной было принято решение пожертвовать этим объявлениями и удалить 1903 строки безвозвратно. Без данных о площади квартиры эти объявления бесполезны. 

```python
data = data.dropna(subset='living_area').reset_index(drop=True)
data.info()
```

Пропуски в столбце 'balcony' заполним нулями. Скорей всего пропуск данных означает отсуствие балкона. В столбце 'is_apartement' пропуски заменим значением False т.к. он содержит логический тип данных (если иное не указано, то скорей всего это обычная квартира). 

```python
data['balcony'] = data['balcony'].fillna(value=0)
```

```python
data['is_apartment'] = data['is_apartment'].fillna(value=False)
```

Пропущенные значения в названии населенного пункта заменим на "Город неизвестен", чтобы 
- не потерять 44 строки,
- не испортить статистику по конкретным городам. 

```python
data['locality_name'] = data['locality_name'].fillna(value='Город неизвестен')
```

```python
data.isna().sum()
```

Посмотрим как можно заменить пропуски в столбце с площадью кухни. Для начала изучим данные.

```python
data['kitchen_area'].describe()
```

Мы наблюдаем распределение Пуассона. Посчитаем сколько в среднем занимает кухня от площади квартиры и заменим отсутствующие значения на общую площадь, умноженную на это число. 

```python
kitchen_proportion = (data['kitchen_area'] / data['total_area']).mean()
kitchen_proportion
```

В среднем кухня занимает почти 1/5 от общей плорщади квартиры. 

```python
data['kitchen_area'] = data['kitchen_area'].fillna(data['total_area'] * kitchen_proportion)
data.info()
```

Исследуем столбец с высотой потолков 'ceiling_height'

```python
data['ceiling_height'].describe()
```

Видим, что высота указана в метрах. При этом, имеются пропуски и аномалии (слишком высокие и слишком низкие потолки). Заполним пропуски и аномалии медианным значением. 

```python
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
```

```python
data.loc[(data['ceiling_height'] > 5) | (data['ceiling_height'] < 2),'ceiling_height'] = data['ceiling_height'].median()
```

```python
data['ceiling_height'].hist()
pass
```

#### Преобразование типов данных

Сразу исправим формат данных у столбца 'first_day_exposition' на специальный формат для дат (это поможет при анализе, если столбец будет использоваться, строковые данные анализировать сложнее)

Также изменим тип данных у столбца 'is_apartment'. Сейчас это строка, а должен быть логический тип данных, что следует из названия столбца.

В датафрейме остался только один столбец со строковыми данными (locality_name), он содержит данные о населенном пункте. Посмотрим, всегда ли один и тот же населенный пункт записывался одинаково (например, есть ли сокращения, использование букв е/ё и т.д.). 

```python
print(len(data['locality_name'].unique()))
print(data['locality_name'].unique())
```

Действительно имеются разные написания одного и того же населенного пункта. Приведём их к единому виду. Начнём с замены буквы ё на е т.к. большинство дубликатов связаны с этим (поселок/посёлок). А также приведём название посёлков городского типа к единому стилю. 

```python
data['locality_name'] = data['locality_name'].replace('ё', 'е', regex=True)
```

```python
data['locality_name'] = data['locality_name'].replace('городской поселок', 'поселок городского типа', regex=True)
```

```python
print(len(data['locality_name'].unique()))
print(data['locality_name'].unique())
```

Удалось сократить дублирующиеся названия на 41 единицу. Теперь обработаем частные случаи

```python
for index, name in data['locality_name'].items():
    if 'Приветнинское' in name:
        data.loc[index, 'locality_name'] = 'поселок Приветнинское'
    elif 'Лесное' in name:
        data.loc[index, 'locality_name'] = 'поселок Лесное'
    elif 'Вещево' in name:
        data.loc[index, 'locality_name'] = 'поселок Вещево'
    elif 'Кудрово' in name:
        data.loc[index, 'locality_name'] = 'деревня Кудрово'
    elif 'Громово' in name:
        data.loc[index, 'locality_name'] = 'поселок Громово'
    elif 'Рябово' in name:
        data.loc[index, 'locality_name'] = 'поселок городского типа Рябово'
```

```python
print(len(data['locality_name'].unique()))
```

### Посчитайте и добавьте в таблицу новые столбцы

Для анализа нам потребуется цена одного квадратного метра. Посчитаем её и занесём в отдельный столбец 'price_per_square_meter'

```python
data['price_per_square_meter'] = data['last_price'] / data['total_area']
```

Добавим столбец с номером дня публикации объявления ('day')

```python
data['day'] = data['first_day_exposition'].dt.weekday
```

Добавим столбец с номером месяца публикации объявления ('month')

```python
data['month'] = data['first_day_exposition'].dt.month
```

Добавим столбец с номером года публикации объявления ('year')

```python
data['year'] = data['first_day_exposition'].dt.year
```

Этаж квартиры тоже влияет на цену. Обычно первые и последние этажи стоят дешевле. Добавим отдельную колонку с типом этажа квартиры (значения — «первый», «последний», «другой»)

```python
def floor_type(floor, floors_total):
    if floor == 1:
        return 'первый'
    elif floor == floors_total:
        return 'последний'
    else:
        return 'другой'
```

```python
data['floor_type'] = data.apply(lambda x: floor_type(x.floor, x.floors_total), axis=1)
```

Для удобства анализа добавим столбец с расстоянием до центра в километрах (и округлим до целых чисел). Посмотрим на итоговый датафрейм после всех преобразований

```python
data['to_city_center'] = round(data['city_center_nearest'] / 1000)
data
```
